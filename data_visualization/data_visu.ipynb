Load Dataframe

```python
import pandas as pd

wines = pd.read_csv("wines.csv",index_col=0)
wines
```

How many wines have been given a rating of 100?

```python
len(wines.loc[wines.points==100])
```

What is the most expensive wine in the dataset?

```python
wines.loc[wines.price == wines.price.max(),'name'].squeeze()
```

Show ratings as grade from 0 to 5

```python
ratio_points_to_grade = 100/5

wines['grades'] = wines['points']/ratio_points_to_grade
wines.grades
```

Create a price histogram for wines with price less than 100

```python
wines.loc[wines.price<100].hist(column='price',bins=20)
```

price vs points

```python
wines.plot(kind='scatter',y='points',x='price',)
```

```python
import seaborn as sns

sns.regplot(y='points', x='price', data=wines.loc[wines.price<100], scatter_kws={"color": "blue"}, line_kws={"color": "red"})
```
